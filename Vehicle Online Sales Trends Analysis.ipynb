# 📌Stage 2
 Data Cleaning and Pre-processing

```python
# Convert datatypes date columns to datetime

df.info()
```

```python
import warnings
warnings.filterwarnings('ignore')

df['makeyear'] = pd.to_datetime(df['makeyear'], errors='coerce')
df['fromdate'] = pd.to_datetime(df['fromdate'], errors='coerce')
df['todate'] = pd.to_datetime(df['todate'], errors='coerce')
df.info()
```

```python
#Handling missing values

numerical_columns = df.select_dtypes(include =['number']).columns
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].median())
print("Numerical columns : \n",numerical_columns)


non_numerical_coulmns = df.select_dtypes(include =  "object").columns
Categorical_col  = df[non_numerical_coulmns].mode().iloc[0]
df[non_numerical_coulmns] = df[non_numerical_coulmns].fillna(Categorical_col)
print("Non_Numerical columns : \n",non_numerical_coulmns)
```

```python
df.dropna(inplace=True)
df.info()
```

```python
df.isnull().sum()
```

```python
# Handling Duplicate - Identify & remove duplicates

df.duplicated().sum()
df.drop_duplicates()
```

```python
# Renaming Columns

df.rename(columns={
    'modeldesc': 'model',
    'vehicleclass': 'vehicle_class',
    'makername': 'maker_name',
    'officecd': 'office_code',
    'vehicle_value_score': 'value_score'
}, inplace=True)

df.columns
```

```python
# Check Skewness

# I used skewness to understand the distribution of numerical data and to identify whether transformation is needed

df['value_score_log'] = np.log1p(df['value_score'])
df['value_score_sqrt'] = np.sqrt(df['value_score'])
print("Skewness:\n", df.skew(numeric_only=True))
```

```python
# Extract year and month

df['year'] = df['fromdate'].dt.year
df['month'] = df['todate'].dt.month

print(df.info())
print(df.head())
```

```python
# Replace -1 in Fuel column:

df['fuel'] = df['fuel'].replace([-1, '-1'], np.nan)
df['fuel'] = df['fuel'].fillna(df['fuel'].mode()[0])

print(df['fuel'].unique())
```

```python
# Add New Column based on Dataset

df['vehicle_age'] = 2026 - df['makeyear'].dt.year
df['vehicle_age'] = df['vehicle_age'].fillna(df['vehicle_age'].median())

df['score_per_seat'] = df['value_score'] / df['seatcapacity']

df.columns
```

#📌Stage 3
## EDA & Visualizations

```python
# 👉 Univariate Analysis - Fuel Type Distribution

colors = sns.color_palette('pastel')

fuel_counts = df['fuel'].value_counts()

explode = [0.01] * len(fuel_counts)

fuel_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=colors,
    explode=explode,
    shadow=True
)

plt.title("Fuel Type Distribution",fontweight="bold")
plt.ylabel('')
plt.show()
```

#**🧠 Interpretation**

*  Petrol vehicles are most common.
*  Limited Diversity in Fuel Choices, Other fuel types like diesel or alternatives have a much smaller proportion.
*  Customers prefer petrol vehicles due to lower cost, easy availability, and maintenance convenience.
*  Manufacturers may currently have a limited portfolio of electric or alternative fuel vehicles.

```python
# Vehicle Class Distribution

plt.figure(figsize=(12,6))

top_classes = df['vehicle_class'].value_counts().head(10)

sns.barplot(x=top_classes.index, y=top_classes.values, palette="Set2", edgecolor="black")
plt.xlabel("Vehicle Class",fontweight="bold")
plt.ylabel("Count",fontweight="bold")
plt.xticks(rotation=45)
plt.title("Distribution of Vehicle Classes",fontweight="bold")
plt.tight_layout()
plt.show()
```

#**🧠 Interpretation**


*   The chart shows that motor cycles dominate the dataset, indicating strong demand for two-wheelers.

*   Other vehicle types have much lower frequency. Shows market is heavily focused toward personal transport vehicles.


```python
# 👉 Bivariate Analysis - Correlation Heatmap

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True,cmap="inferno",linewidths=0.5)
plt.title("Correlation Matrix",fontweight="bold")
plt.show()
```

#**🧠 Interpretation**

*   Seat capacity vs Value Score (~0.35)
→ Moderate positive relationship
→ Vehicles with more seats tend to have slightly higher value

*   Month correlation (~0)
→ No strong relationship with value or seat capacity
→ Indicates no strong seasonal impact

*  Near-zero correlations overall
→ Most variables are independent
→ No strong linear relationship exists












```python
# BoxPlot - Impact of Secondhand Vehicle

plt.figure(figsize=(10,6))

sns.boxplot(
    data=df,
    x='secondvehicle',
    y='value_score',
    hue='fuel'
)

plt.title('Impact of Second Vehicle Ownership')
plt.show()
```

# **🧠 Interpretation**



*   Each box shows distribution per fuel type
insights:
Battery → wider spread → more variability
Diesel → more consistent
CNG → may show lower median

*   People owning a second vehicle tend to choose higher value vehicles.

*   Outliers in both groups indicate the presence of extreme high and low-value vehicles, suggesting that exceptional cases are not strongly dependent on ownership status.





```python
# Manufacturer vs Vehicle Count (Top 15)

plt.figure(figsize=(10,5))

top_makers = df['maker_name'].value_counts().head(15)

sns.barplot(
    x=top_makers.values,
    y=top_makers.index,
    palette='viridis',
    edgecolor='black'
)

plt.title("Top 15 Manufacturers", fontweight='bold')
plt.xlabel("Count")
plt.ylabel("Manufacturer")

plt.tight_layout()
plt.show()
```

*   Customers tend to prefer well-known and established brands

```python
# 👉 Multivariate Analysis - Pairplot

sns.pairplot(df, hue='fuel')
plt.suptitle("Pairplot - Multivariate Analysis", fontweight='bold')
plt.show()
```

#**🧠 Interpretation**

```python
# Plotly - Histogram

fig = px.scatter_3d(
    df,
    x='vehicle_age',
    y='value_score',
    z='seatcapacity',
    color='fuel',
    symbol='vehicle_class',
    title='3D Multivariate Analysis',
    template='plotly_dark'
)
fig.show()
```

#**🧠 Interpretation**


*   High seat capacity vehicles are very few and these vehicles do not show consistently high value scores

*   There is clear clustering based on vehicle type, separating personal vehicles from commercial/high-capacity ones.

*   Diesel vehicles are more associated with higher-capacity segments, while petrol vehicles dominate low-capacity personal vehicles.


```python
# Pivot Table Analysis


pivot = df.pivot_table(
    values='value_score',
    index='month',
    columns='fuel',
    aggfunc='mean'
)
pivot.fillna(0, inplace=True)
pivot

```

# **🧠 Interpretation**

*  Diesel vehicles consistently maintain higher value scores across all months, indicating strong market preference and stability.

*   CNG-based vehicles show inconsistent trends due to limited data, making them less reliable for pattern analysis.

*   Battery vehicles show moderate and relatively stable value scores, indicating gradual adoption in the market.

*   Overall, fuel type plays a significant role in determining vehicle values.
